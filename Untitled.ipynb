# Boston Crime Data Analysis

### Importing all the neccessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

### Importing CSV files of the datasets

### Checking and removal of NaN or missing values

#### We are going to be removing NaN or missing values from different columns of the dataset 

```python
print("Total number of NaN values is column SHOOTING :",df['SHOOTING'].isnull().sum())
```

#### Replacing NaN values with 'N'

```python
df['SHOOTING'].replace(np.nan,'N',inplace = True)
```

#### Confirming the replacement of NaN values in the required column

```python
print("Total number of NaN values is column SHOOTING :",df['SHOOTING'].isnull().sum())
```

### Exploratory Data Analysis

#### Number of offenses for each type of offense code

```python
no_of_offenses = df['OFFENSE_CODE'].value_counts()
no_of_offenses.to_frame('No of crimes').head()
```

#### Combining the two datasets

```python
df_combine = pd.merge(df,df_code,on='OFFENSE_CODE',how='inner')
df_combine.head()
```

#### Number of crimes with crime name 

```python
no_of_crime_off = df_combine['NAME'].value_counts()
no_of_crime_off.to_frame('Number of Crimes').head(20)
```

#### Number of crimes each year

```python
no_of_crime_yrs = df['YEAR'].value_counts()
no_of_crime_yrs.to_frame('Number of Crimes').head()
```

```python
ax = no_of_crime_yrs.plot.bar(rot=0)
plt.xlabel('Years')
plt.ylabel('Number of Crimes')
```

#### Number of crimes according to Offense Code Group

```python
no_of_crime_cd_grp = df['OFFENSE_CODE_GROUP'].value_counts()
no_of_crime_cd_grp.to_frame('Number of Crimes').head(25)
```

#### Number of crimes at each street

```python
no_of_crime_street = df['STREET'].value_counts()
no_of_crime_street.to_frame('Number of Crimes').head(20)
```

#### Boxplot for shootings over the years
